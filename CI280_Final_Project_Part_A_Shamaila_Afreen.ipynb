```python
# 3-4) Show the death rate in United Kingdom , 
# Is it above or lower than the worldwide death rate?
# (3 pts)

number = random.randint(0,10000)
print("Shamaila Afreen, 21500", number )
print('------------------------------')

res = df[df['location']=='United Kingdom']
num1 = res['new_cases'].sum()


res = df[df['location']=='United Kingdom']
num2 = res['new_deaths'].sum()

rate1 = num2/num1
print('Death rate in United Kingdom: ', format(rate1, '.2%'))

rate2 = df['new_deaths'].sum()/df['new_cases'].sum()
print('Death rate in the World: ', format(rate2, '.2%'))

if rate1 > rate2:
    print('United Kingdom is above the world level')
else:
    print('United Kingdom is below the world level')


```

```python
# 5-6) Show the average new_cases_per_million people in United Kingdom
# Is it above or lower than the worldwide level.
# (6 pts)
number = random.randint(0,10000)
print("Shamaila Afreen, 21500", number )
print('------------------------------')

res = df[df['location']=='United Kingdom']
num1 = res['new_cases_per_million'].mean()
print('Average new cases per million in the United Kingdom: ', format(num1, '.0f'))

num2 = df['new_cases_per_million'].mean()
print('Average new cases per million in the world: ', format(num2, '.0f'))

if num1> num2:
    print('United Kingdom is above the world level')
else:
    print('United Kingdom is below the world level')
```
